```python
%%sql
select
orderdate,orderkey,linenumber,
orderkey*10+linenumber as order_line_number,
(exchangerate*netprice*quantity) as net_revenue,
sum(exchangerate*netprice*quantity) over(partition by orderdate) as daily_net_revenue,
(exchangerate*netprice*quantity)*100/sum(exchangerate*netprice*quantity) over(partition by orderdate) as pct_daily_revenue
from sales
order by
  orderdate,
  pct_daily_revenue desc
limit 10
```

```python
%%sql
SELECT *,
100 * net_revenue / daily_net_revenue AS pct_daily_revenue
FROM (
SELECT
orderdate,
orderkey * 10 + linenumber AS order_line_item, (quantity * netprice * exchangerate) AS net_revenue,
SUM(quantity * netprice * exchangerate) OVER(PARTITION BY orderdate) AS daily_net_revenue
FROM
sales
) AS revenue_by_day
```

```python
%%sql
with yearly_cohort as(
SELECT distinct customerkey,
extract(YEAR from min(orderdate) over(partition by customerkey)) as cohort_year
from sales
)
select y.cohort_year,
extract(year from orderdate) as purchase_year ,
SUM(quantity * netprice * exchangerate)as net_revenue
from sales s
left join yearly_cohort y on s.customerkey=y.customerkey
group by y.cohort_year,purchase_year



```

```python
%%sql
with yearly_cohort as(
select distinct customerkey,
extract(year from min(orderdate) over(partition by customerkey)) as cohort_year,
extract(year from orderdate) as purchase_year
from sales
)
select distinct cohort_year,
purchase_year,count(customerkey) over(partition by cohort_year,
purchase_year) as num_customers
from yearly_cohort
order by cohort_year,
purchase_year
```

```python
%%sql
with monthly_revenue as(
select
  to_char(orderdate,'yyyy-mm') AS order_month,
  sum(quantity*netprice*exchangerate) as net_revenue
from sales
where extract(year from orderdate)='2023'
--where orderdate between '2023-01-01' and '2023-12-31'
group by order_month
order by order_month
)
select*,
        first_value(net_revenue) over(order by order_month) as  first_month_revenue,
        last_value(net_revenue) over(order by order_month rows between unbounded preceding and unbounded following) as  last_month_revenue,
        nth_value(net_revenue,3) over(order by order_month rows between unbounded preceding and unbounded following) as  third_month_revenue
from monthly_revenue

```

```python
%%sql
with monthly_revenue as(
select
  to_char(orderdate,'yyyy-mm') AS order_month,
  sum(quantity*netprice*exchangerate) as net_revenue
from sales
where extract(year from orderdate)='2023'
--where orderdate between '2023-01-01' and '2023-12-31'
group by order_month
order by order_month
)
select*,
        lag(net_revenue) over(order by order_month) as  previous_month_revenue,
        lag(net_revenue,3) over(order by order_month) as  previous_month_revenue,--3 previous
        lead(net_revenue) over(order by order_month) as  next_month_revenue,
        lead(net_revenue,3) over(order by order_month) as  next3_month_revenue--3 month front
from monthly_revenue
```

```python
%%sql
with monthly_revenue as(
select
  to_char(orderdate,'yyyy-mm') AS order_month,
  sum(quantity*netprice*exchangerate) as net_revenue
from sales
where extract(year from orderdate)='2023'
--where orderdate between '2023-01-01' and '2023-12-31'
group by order_month
order by order_month
)
select*,
        lag(net_revenue) over(order by order_month) as  previous_month_revenue,
        net_revenue-lag(net_revenue) over(order by order_month) as  month_rev_growth,--diff
        100*(net_revenue-lag(net_revenue) over(order by order_month))/lag(net_revenue) over(order by order_month) as  month_rev_growth_percent
from monthly_revenue
```

```python
%%sql
with monthly as(
select
extract(year from min(orderdate)) as cohort_year,
sum(quantity*netprice*exchangerate) as sum_lifetime
from sales
group by customerkey
),
avg_cohort as(
select cohort_year,avg(sum_lifetime)as avg_lifetime
from monthly
group by cohort_year
)
select *,
lag(avg_lifetime)over(order by cohort_year) as prev_avg_lifetime,
100*(avg_lifetime-lag(avg_lifetime)over(order by cohort_year))/lag(avg_lifetime)over(order by cohort_year)as percent_change,
lead(avg_lifetime)over(order by cohort_year) as lead_avg_lifetime
from avg_cohort
order by cohort_year

```
